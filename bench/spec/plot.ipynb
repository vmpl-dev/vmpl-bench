```python
import os
import re
import sys

import matplotlib.pyplot as plt

def plot_benchmark_results(home_dir, file="CINT2006.001.txt"):
    # Open the file
    filepath = os.path.join(home_dir, "result", file)
    with open(filepath, 'r') as f:
        content = f.read()

    # Use a regular expression to extract benchmark results
    pattern = r"(\d{3}\.\w+)\s+\d+\s+(\d+)\s+([\d\.]+)"
    results = re.findall(pattern, content)

    # Separate the results into names, run times, and rates
    names, run_times, rates = zip(*results)

    # Convert run times and rates to integers and floats
    run_times = list(map(int, run_times))
    rates = list(map(float, rates))

    # Create a bar chart for the run times
    f = plt.figure(figsize=(10, 5))
    plt.bar(names, run_times)
    plt.xlabel('Benchmark')
    plt.ylabel('Run Time')
    plt.title('Benchmark Run Times')
    plt.xticks(rotation=30)
    plt.tight_layout()
    plt.show()
    f.savefig(f"{file}-time.pdf")

    # Create a bar chart for the rates
    f = plt.figure(figsize=(10, 5))
    plt.bar(names, rates)
    plt.xlabel('Benchmark')
    plt.ylabel('Rate')
    plt.title('Benchmark Rates')
    plt.xticks(rotation=30)
    plt.tight_layout()
    plt.show()
    f.savefig(f"{file}-rate.pdf")

SPEC2006="/home/benshan/speccpu/speccpu2006/speccpu2006-v1.0.1"

# 2024年3月28日01:49:00
plot_benchmark_results(SPEC2006, "CINT2006.001.txt")
plot_benchmark_results(SPEC2006, "CFP2006.001.txt")

# 2024年4月2日14:13:13
# benshan@sev-snp-guest:~/speccpu/speccpu2006/speccpu2006-v1.0.1$ runspec -c x86.cfg -n 1 -r 1 all
plot_benchmark_results(SPEC2006, "CINT2006.010.txt")
plot_benchmark_results(SPEC2006, "CFP2006.010.txt")
```
